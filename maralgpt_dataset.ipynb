```python
import pandas as pd
from datasets import Dataset, load_dataset
from huggingface_hub import notebook_login
```

```python
notebook_login()
```

```python
dataset = load_dataset("sinarashidi/alpaca-persian", split="train")
```

```python
df = dataset.to_pandas()
df.head()
```

```python
data = []
```

```python
for row in df.iterrows():

  text = row[1]['text']
  text = text.split("### Human:")
  text = text[1]
  text = text.split("### Assistant:")
  text = f"<|USER|>\n{text[0]}\n<|ENDOFUSER|>\n<|ASSISTANT|>\n{text[1]}\n<|ENDOFASSISTANT|>"

  datum = {}
  datum['text'] = text

  data.append(datum)
```

```python
print(data[10]['text'])
```

```python
new_df = pd.DataFrame(data)
new_df.head()
```

```python
new_data = Dataset.from_pandas(new_df)
```

```python
new_data
```

```python
new_data.push_to_hub("MaralGPT/maralgpt-dataset-v0-1")
```
